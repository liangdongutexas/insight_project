In the above method, I considered put cpu, gpu bench mark as input and predict fps given a game. In the majority, fps goes up with benchmark. However, there are several cases that as the bench mark goes up, the fps goes down. This exceptional cases may be due to 
1. factors associated with brand dependence between AMD and Intel
2. architecture change in the design of cpu,gpu.
3. fps not only depends on benchmark, also depends on details such as number of threads, cores and frequency. The detailed relation can be complicated. 

These exceptional cases rules out the idea of simply using benchmark as input. I can choose cpu, gpu details but the number of input goes up so the parameters in a model.
This may still be viable by choosing data from different games.

## Building a model to predict FPS

From the above example, it is clear that for any two cpu-gpu combinations, the FPS increment is linear between each other. 
So we can guess a model as

\begin{align} 
F^{i}_{mn}=g^{i}P_{mn}+\alpha_{mn}
\end{align}
where $i$ is the label for games, $mn$ are the label for gpu and cpu respectively, and $\alpha$ contains other information that is game independent.

Next, I will use the current data to testify this model.

The number of parameters in this model is $i+2*m*n$.
Because we find 'i' games fps benchmark, the number of data point will be $i*m*n$.
In the case where $i=24,m=28,n=14$,the number of parameters are $808$ and the number of data points is $9408$.

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
## find2d returna a 2d array representing fps among 
## different cpu, gpu combinations given a game
def find2d(game_name):
    game_fps=pd.read_csv(game_name+'.csv')
    game_fps=np.array(game_fps)
    game_fps=[i[1:] for i in game_fps]
    game_fps=np.array(game_fps).astype(np.float32)
    return game_fps
```

```python
game_names=pd.read_csv('games_fps_hyperlinks.csv')['game_name']
```

```python
fps_all=np.array([find2d(i) for i in game_names])
```

```python
fps_all.shape
```

The shape of fps_all file is (24,28,14). 24 is the number of games tested. 28 is the number of gpus tested and 14 is the number of cpus tested.

\begin{align} 
F^{i}_{mn}=g^{i}P_{mn}+\alpha_{mn}
\end{align}

```python
## Here I will use tensor flow to do it.
class model():
    def __init__(self):
        self.g=tf.Variable(tf.random.truncated_normal(shape=(24,)))
        self.P=tf.Variable(tf.random.truncated_normal(shape=(28,14)))
        self.alpha=tf.Variable(tf.random.truncated_normal(shape=(28,14)))
        self.trainable_variables=[self.P,self.alpha,self.g]
    def __call__(self):
        F_predict=tf.concat([tf.expand_dims(self.g[i]*self.P,0) for i in range(24)],0)\
                    +tf.tile(tf.expand_dims(self.alpha,0),[24,1,1])
        return F_predict
    def load_variables(self,P,alpha,g=None):
        if g:
            self.g=tf.constant(g)
        self.P=tf.constant(P)
        self.alpha=tf.constant(alpha)
        
```

```python
def compute_loss(F_predict, F):
    mse = tf.keras.losses.MeanSquaredError()
    return mse(F_predict,F)
```

```python
def train_one_step(model,F):
    with tf.GradientTape() as tape:
        F_predict = model()
        loss = compute_loss(F_predict,F) 
        # compute gradient
        grads = tape.gradient(loss, model.trainable_variables)
        # update to weights
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
```

```python
def build(model,epochs,F):   
    for epoch in range(epochs):           
        train_one_step(model,F)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
testmodel=model()
```

```python
epochs=300
build(testmodel,epochs,fps_all)
```
